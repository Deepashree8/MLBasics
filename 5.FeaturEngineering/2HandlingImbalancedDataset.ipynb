Handling Imbalanced Dataset 

1.Up Sampling 
2.Down Sampling 

```python
import pandas as pd
import numpy as np
##set the random seed for reproducibility 
np.random.seed(123)

#create a dataframe with two classes
n_samples=1000
class_0_ratio=0.9
n_class_0=int(n_samples*class_0_ratio)
n_class_1=n_samples-n_class_0
```

```python
n_class_0,n_class_1
```

```python
##creation Dataframe with Imbalanced Dataset 

# Generates n_class_0 random numbers
# Follows a normal (Gaussian) distribution
# Mean (loc) = 0
# Standard deviation (scale) = 1
# Example values: -0.2, 0.4, 1.2, -1.1…
class_0=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})
class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
##upsampling 
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
df_majority
```

```python
df_minority
```

```python
##resample by taking the minority data,replace=tRUE can pick row multiple times
#random_state=42 everytime gives same result
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True,
                               n_samples=len(df_majority),
                               random_state=42)
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```
