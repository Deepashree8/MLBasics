Data Encoding 
1.Nominal/OHE encoding=One-Hot Encoding is a technique that converts categorical values into numerical format by creating a separate binary column for each unique category. Each category is represented as 1 (present) or 0 (absent).


2.Label and ordinal encoding 


3.Target Guided ordinal Encoding 

```python
##1.Nominal/OHE Encoding 
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder 
```

```python
##creation of simple dataframe
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df['color']
```

```python
df.head()
```

```python
#instances of OneHotEncoder 
encoder=OneHotEncoder()
```

```python
##perform fit and transform 
encoded=encoder.fit_transform(df[['color']]).toarray()
encoded
```

```python
import pandas as pd 
encoded_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
##for new data 
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoded_df],axis=1)
```

Label Encoding is a technique that converts categorical text values into numerical labels by assigning each unique category an integer (e.g., red → 0, blue → 1),labels assigned in alphabetical order or based on frequency of the categories

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder 
lbl_encoder=LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

OrdinalEncoding:
encode categorical data that have an instrinctic order or ranking.
each category is assigned a numerical value based on its position in the order

```python
##ordinal encoding 
from sklearn.preprocessing import OrdinalEncoder
```

```python
#creation of sample dataframe
df=pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
#creation of instances of ordinal encoder 
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```
