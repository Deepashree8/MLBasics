MAE — Mean Absolute Error

It measures the average absolute difference between predicted and actual values.

Interpretation: Lower MAE = better model; error measured in the same units as the target.

RMSE — Root Mean Squared Error

It measures the square root of the average squared errors.


Interpretation:

RMSE gives more penalty to large errors (because errors are squared).

Good for cases where large mistakes are costly.

MAPE — Mean Absolute Percentage Error

It measures error in percentage form.
Interpretation:

Gives how far predictions are from actual values in %.

Example: MAPE = 8% → predictions are 8% off on average.

Limitation: Can’t be used when actual value is 0.

4. Time-Based Validation (Train-Test Split for Time Series)

For time series, data must be split in chronological order, not randomly.

Purpose: ensures the model predicts future values using only past data.

Example:

Train: Jan–Sep

Test: Oct–Dec

Prevents data leakage and mimics true forecasting.

Concept:

Time series models must respect the time dependency, so validation is done using forward chaining, rolling windows, or expanding windows.

Forward Chaining (Walk-Forward Validation)

Concept:

You train on past data and test on the next future data point(s).

After each step, the training set expands, and the test moves forward.

Ensures predictions always use only past → future, just like real forecasting.

Example:

Train: [1,2,3]   | Test: [4]
Train: [1,2,3,4] | Test: [5]
Train: [1,2,3,4,5] | Test: [6]


Rolling Window (Sliding Window Validation)

Concept:

Training window size is fixed.

When new data enters, the oldest data point is removed.

Window “slides” forward through time.

Example (window = 3):

Train: [1,2,3] | Test: [4]
Train: [2,3,4] | Test: [5]
Train: [3,4,5] | Test: [6]




Expanding Window (Growing Window)

Concept:

Similar to forward chaining, but the train window keeps growing indefinitely.

You never drop old data; you keep expanding your training set.

Example:

Train: [1,2,3] | Test: [4]
Train: [1,2,3,4] | Test: [5]
Train: [1,2,3,4,5] | Test: [6]


Why?

Regular train_test_split (from sklearn):

randomly splits data

breaks time order → NOT allowed for time series

So for time series, we need special methods that respect chronological order:






Time-Series Train–Test Split Methods

Expanding Window → train keeps growing

Forward Chaining → multiple expanding splits

Rolling Window → fixed window slides forward

Blocked Time Split → simple past → future split

TimeSeriesSplit in sklearn → implements forward chaining
