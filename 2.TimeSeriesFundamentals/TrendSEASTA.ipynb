Trend is the long-term upward or downward movement in time-series data.
It shows the overall direction of the data over a long period.

Why trend occurs?

1.Economic growth
2.Increase in population
3.Business expansion
4.Technological improvement

How to detect trend?

1.Line plot
2.Rolling mean
3.Decomposition (statsmodels)

```python
##identifying trend
import pandas as pd
import matplotlib.pyplot as plt 
##sample data with trend
data={
    "month":pd.date_range(start="2024-01-01",periods=36,freq="ME"),
    "sales":[200,220,230,240,260,280,300,310,330,350,370,400,
             420,440,460,470,490,510,530,550,570,600,620,650,
             670,690,710,730,750,770,800,820,850,870,900,920]
}
df=pd.DataFrame(data)
df.head(4)
df.set_index("month",inplace=True)
```

```python
df.head(4)
```

```python
plt.plot(df.sales)
plt.title("Trend")
```

Seasonality is a pattern that repeats at regular intervals (daily, monthly, yearly).
Example:

1.Ice cream sales increase every summer
2.Electricity usage increases every evening
3.Online shopping spikes during festivals
4.Seasonality is fixed and predictable.
5.Frequency could be:
12 months (monthly data)
7 days (daily data)
24 hours (hourly data)

time series decomposition, which means it breaks a time series into 3 components:
1. Trend
The long-term movement of data over time
(Example: sales increasing each year)
2. Seasonality
Repeating patterns every fixed period
(Example: sales high every December)
3. Residual / Noise
Random fluctuations
(Example: unexpected high or low sales)

4.model=additive ,additive decomposition (changes are linear),changes large over time multiplicative
time series=Trend+Seasonality+Residual

```python
##Identifying seasonality using seasonal_decompose
import pandas 
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition=seasonal_decompose(df['sales'],model="additive",period=12)
decomposition.plot()
plt.show()

```

STATIONARY

A time series is stationary if:

1.Mean is constant
2.Variance is constant
3.No seasonality
4.No trend

Stationary data is REQUIRED for many time-series models like ARIMA.
Non-stationary Examples
1.Data with trend
2.Data with seasonality

Stationary Examples:
Noise-like data with no pattern 
Mean fluctuations around a constant line

ADF (Augmented DickeyFuller) test is a statistical test used to check whether a time series is stationary or not.
It tests if the series has a unit root (non-stationarity).
If the p-value < 0.05, the series is considered stationary.

```python
##check stationary(ADF Test)
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Example time series
data = [120, 123, 125, 128, 130, 129, 131, 133, 132, 134]
ts = pd.Series(data)

# ADF Test
result = adfuller(ts)

print("ADF Statistic:", result[0])
print("p-value:", result[1])

if result[1] < 0.05:
    print(" The series is stationary")
else:
    print(" The series is NOT stationary")
```
