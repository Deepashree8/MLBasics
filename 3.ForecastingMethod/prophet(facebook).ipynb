Prophet is a robust time series model that handles:

Trend (linear or logistic)
Seasonality (daily, weekly, yearly)
Holidays and special events

It is very user-friendly and works well with missing data or outliers.

```python
from prophet import Prophet 
import pandas as pd
import matplotlib.pyplot as plt

# Example: Monthly sales data
data = pd.Series([112,118,132,129,121,135,148,148,136,119,104,118,
                  115,126,141,135,125,149,170,170,158,133,114,140], 
                 index=pd.date_range(start='2020-01', periods=24, freq='ME'))
#prepare data for Prophet 
df=pd.DataFrame({'ds':data.index,'y':data.values})

#Fit Prophet model
prophet_model=Prophet(yearly_seasonality=True,daily_seasonality=False)
prophet_model.fit(df)

#Make future dataframe for 12 months 
future=prophet_model.make_future_dataframe(periods=12,freq='ME')
forecast=prophet_model.predict(future)
#plot the forecast 
prophet_model.plot(forecast)
plt.show()
```
